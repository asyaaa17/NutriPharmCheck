```python
def replace_capsule(text):
    if '1 капс.' in text:
        text = text.replace('1 капс.', '1 капсуле')
    if '2 капс.' in text:
        text = text.replace('2 капс.', '2 капсулы')
    if '3 капс.' in text:
        text = text.replace('3 капс.', '3 капсулы')
    if '4 капс.' in text:
        text = text.replace('4 капс.', '4 капсулы')
    if '5 капс.' in text:
        text = text.replace('5 капс.', '5 капсул')
    if '6 капс.' in text:
        text = text.replace('6 капс.', '6 капсул')
    if '7 капс.' in text:
        text = text.replace('7 капс.', '7 капсул')
    if '8 капс.' in text:
        text = text.replace('8 капс.', '8 капсул')
    if '10 капс.' in text:
        text = text.replace('10 капс.', '10 капсул')
    if '(капс.)' in text:
        text = text.replace('(капс.)', '(капс)')
    return text
```

```python
df['Рекомендации по употреблению'] = df['Рекомендации по употреблению'].progress_apply(replace_capsule)
```

```python
def replace_pill(text):
    if '1 табл.' in text:
        text = text.replace('1 табл.', '1 таблетке')
    if '2 табл.' in text:
        text = text.replace('2 табл.', '2 таблетки')
    if '3 табл.' in text:
        text = text.replace('3 табл.', '3 таблетки')
    if '4 табл.' in text:
        text = text.replace('4 табл.', '4 таблетки')
    if '5 табл.' in text:
        text = text.replace('5 табл.', '5 таблеток')      
    if '6 табл.' in text:
        text = text.replace('6 табл.', '6 таблеток')         
    if '7 табл.' in text:
        text = text.replace('7 табл.', '7 таблеток')
    if '8 табл.' in text:
        text = text.replace('8 табл.', '8 таблеток')
    if '½ табл.' in text:
        text = text.replace('½ табл.', '½ таблетки')
    if '(табл.)' in text:
        text = text.replace('(табл.)', '(табл)')
    if '8 гран.' in text:
        text = text.replace('8 гран.', '8 гранул')    
    return text
```

```python
def replace_month(text):
    if '1 мес.' in text:
        text = text.replace('1 мес.', '1 месяц.')
    if '1мес.' in text:
        text = text.replace('1мес.', '1 месяц.')
    if '2 мес.' in text:
        text = text.replace('2 мес.', '2 месяца.')
    if '3 мес.' in text:
        text = text.replace('3 мес.', '3 месяца.')  
    if '4 мес.' in text:
        text = text.replace('4 мес.', '4 месяца.') 
    if '1,5 мес.' in text:
        text = text.replace('1,5 мес.', '1,5 месяца.')
    return text
```

```python
def num_splits(num, foo):
    df.loc[df.length == num, 'splitted'] = df.loc[df.length == num, 'splitted'].progress_apply(foo)
    df.loc[df.length == num, 'age_dose'] = df[df.length == num]['splitted'].apply(lambda x: x[0])
    df.loc[df.length == num, 'duration'] = df[df.length == num]['splitted'].apply(lambda x: x[1])
    df.loc[df.length == num, 'repeat'] = df[df.length == num]['splitted'].apply(lambda x: x[2])
    return df
```
